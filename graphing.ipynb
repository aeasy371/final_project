```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
mnist_lines = []
with open('cifar100_training.txt', 'r') as f:

	mnist_lines = f.readlines()
```

```python
training_losses = []
training_accuracies = []
test_losses = []
test_accuracies = []
for line in mnist_lines:
	if 'Epoch' in line:
		l = line.split(':')[1].split()
		training_losses.append(float(l[1]))
		training_accuracies.append(float(l[3]))
	elif 'Test loss' in line:
		l = line.split(':')
		test_losses.append(float(l[1].split('%')[0]))
	elif 'Test Accuracy' in line:
		l = line.split(':')
		test_accuracies.append(float(l[1].split('%')[0]))
# training_accuracies
```

```python
training_accuracies = np.array(training_accuracies) * 100
test_accuracies = np.array(test_accuracies)
plt.plot(training_accuracies, c='red', label='Train')
plt.plot(test_accuracies, c='blue', label='Validation')
plt.legend()
plt.title('CIFAR-100 Training vs Validation Accuracy')
```

```python
training_accuracies = np.array(training_losses)
test_accuracies = np.array(test_losses)
plt.plot(training_accuracies, c='red', label='Train')
plt.plot(test_accuracies, c='blue', label='Validation')
plt.legend()
plt.title('CIFAR-100 Training vs Validation Losses')
```
